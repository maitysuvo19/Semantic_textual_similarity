```python
from sentence_transformers import SentenceTransformer, util
import numpy as np
import pandas as pd
```

```python
model = SentenceTransformer('stsb-roberta-large')
```

```python
#loading the data set
df=pd.read_csv('/content/drive/MyDrive/nlp data/Text_Similarity_Dataset.csv')
```

```python
#view first few rows of the dataset
df.head(3)
```

```python
text1=df.text1.tolist()
text2=df.text2.tolist()
```

```python
similarity_score=[]
for index, row in df.iterrows():
  sentence1=text1[index]
  sentence2=text2[index]
  embedding1 = model.encode(sentence1, convert_to_tensor=True)
  embedding2 = model.encode(sentence2, convert_to_tensor=True)# compute similarity scores of two embeddings
  cosine_scores = util.pytorch_cos_sim(embedding1, embedding2)
  similarity_score.append(cosine_scores.item())
```

```python
similarity_score
```

```python
#transforming similarity score sclae [-1,1] to [0,1]
Similarity_Score = [((x+1)/2) for x in similarity_score]
```

```python
df=df.assign(Similarity_Score=Similarity_Score)
df.head(5)
```

```python
df=df[['Unique_ID','Similarity_Score']]
df.head()
```

```python
from google.colab import files
df.to_csv('df_transformer_sts.csv',index=False)
files.download('df_transformer_sts.csv')
```
